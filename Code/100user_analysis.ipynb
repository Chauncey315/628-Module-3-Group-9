```python
index=[]
for i in range(len(df_category_user)):
    score=[]
    score.append(df_category_user['Chinese'][i])
    score.append(df_category_user['Japanese'][i])
    score.append(df_category_user['Korean'][i])
    score.append(df_category_user['Thai'][i])
    score.append(df_category_user['Vietnamese'][i])
    rank = [index for index,value in sorted(list(enumerate(score)),reverse=True,key=lambda x:x[1])]
    index.append(rank[0])
```

```python
Label=[]
for oneindex in index:
    if oneindex==0:
        Label.append('Chinese')
    if oneindex==1:
        Label.append('Japanese')
    if oneindex==2:
        Label.append('Korean')
    if oneindex==3:
        Label.append('Thai')
    if oneindex==4:
        Label.append('Vietnamese')
  
```

```python
df_category_user['Label']=Label
```

```python
newcategory_user=df_category_user[['user_name','user_id','Label']]
```

```python
newcategory_user.to_csv('Label_user.csv', index=False)
```
